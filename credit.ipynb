```python
import numpy as np
import pandas as pd

import matplotlib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
#loading the data set to pandas data frame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first 3 names of the dataset
print(credit_card_data.head(3))
```

```python
#seperating data for analaysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Data is unbalanced so under sampling method is needed to use.

building a sample dataset containing similar distribution for normal transaction and fraudulent transaction

no of fraudelent = 492

```python
legit_sample=legit.sample(n=492)
```

concatinating two data frames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splitting the variable into feature sand target


```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the dataset in training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model=LogisticRegression()
```

```python
#training the logistic regression model with training data
model = make_pipeline(StandardScaler(), LogisticRegression())
model.fit(X_train, Y_train)
Pipeline(steps=[('standardscaler', StandardScaler()),
                ('logisticregression', LogisticRegression())])
model.score(X_test, Y_test)
```

```python
model.score(X_test, Y_test)
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
#print accuracy on training data
print('Accuracy on training data:',training_data_accuracy)
```

```python
#accuracy for prediction of test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on testing data:',test_data_accuracy)
```
